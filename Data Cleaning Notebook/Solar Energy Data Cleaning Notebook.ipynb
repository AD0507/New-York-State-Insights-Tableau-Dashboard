```python
import pandas as pd
import numpy as np

#reading the CSV file containing solar energy data into a pandas DataFrame
df=pd.read_csv('/Users/ad0507/Downloads/VAD Project/Dataset Uncleaned/Solar Energy.csv')

#displaying the first 5 rows of the DataFrame to preview the data
df.head()
```

```python
#check for nulls and also the columns we have
df.info()
```

```python
#define a threshold for dropping columns (e.g., drop columns with more than 50% null values)
threshold = 0.5
columns_to_keep = df.columns[df.isnull().mean() <= threshold]

#drop columns with more than 50% null values
filtered_data = df[columns_to_keep]

#check the shape of the dataset after dropping columns
filtered_data_shape = filtered_data.shape

#display the dataset
filtered_data_shape
```

```python
filtered_data.info()
```

```python
#check the statistics for the columns
filtered_data.describe()
```

```python
#save the merged dataframe to a CSV file for Tableau
merged_output_path = "/Users/ad0507/Downloads/Solar Energy/Solar_Energy_Cleaned.csv"
filtered_data.to_csv(merged_output_path, index=False)
```
