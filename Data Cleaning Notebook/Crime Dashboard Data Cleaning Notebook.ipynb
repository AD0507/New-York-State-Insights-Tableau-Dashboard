```python
import pandas as pd
import numpy as np

#load the dataset
file_path = '/Users/ad0507/Downloads/VAD Project/Dataset Uncleaned/Index_Crimes_by_County_and_Agency.csv'
data = pd.read_csv(file_path)
data.head(20)
```
